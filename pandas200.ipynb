```python
# Initialize an empty list to store results
result1_data = []

# Iterate over columns using iteritems
for column_name, column_data in judge_df.items():
    # Count missing values in each column
    missing_values = column_data.isna().sum()
    
    # Append results to the list
    result1_data.append({'Column': column_name, 'Total Missing Values': missing_values})

# Create a new DataFrame from the results
result1_df = pd.DataFrame(result1_data)

# Display the result DataFrame
print(result1_df)
```

```python
# Creating a data frame to find the total missing values in each columns
# initialize an empty list
result_data2 = pd.DataFrame()
result_data2["Col_Name"]=judge_df.columns
missing_values_list = []
for column in judge_df.columns:
    # Count missing values in each column
    missing_values = judge_df[column].isna().sum()
    missing_values_list.append(missing_values)

result_data2["Total_Missing_Values"] = missing_values_list
# Display the result DataFrame
print(result_data2) 
```

```python
judge_df.filter(["pitch_type","release_pos_z"]).\
    assign(release_pos_z_mean = judge_df["release_pos_z"]/judge_df["release_pos_z"].mean(),
           release_pos_z_z = (judge_df["release_pos_z"]-judge_df["release_pos_z"].mean())/
           judge_df["release_pos_z"].std())
```

```python
judge_df.filter(["pitch_type","stand","plate_x"]).\
    query("pitch_type in ['FF','CH','SL']").\
        assign(plate_x_up = (judge_df['plate_x']+judge_df['plate_x'].mean())/judge_df['plate_x'].std())
```

```python
# assign with regex on des
judge_df.filter(['pitch_type','des','plate_x']).\
    query('~(plate_x >1)').\
        assign(des_new=judge_df['des'].str.extract('(Aaron)'))
```

```python
# # assign with regex on des for Aaron and Tony with out the query 
pattern = '(Aaron|Tony)'
judge_df.filter(['pitch_type','des','plate_x']).\
        assign(des_new=judge_df['des'].str.extract(pattern)).\
            query('des_new in ["Aaron","Tony"]').\
                groupby(['des_new']).\
                    size()
```

```python
# use of reset index + custom function to find unique
find_unique = lambda x:pd.unique(x)
# with reset index
judge_df.filter(['pitch_type','home_team','away_team']).\
    groupby(['pitch_type']).\
        agg({'home_team':find_unique, 'away_team':find_unique}).\
            reset_index()
```

```python
# without reset index
judge_df.filter(['pitch_type','home_team','away_team']).\
    groupby(['pitch_type']).\
        agg({'home_team':find_unique, 'away_team':find_unique})
```

```python
# total unique count : nunique
judge_df.filter(['pitch_type','home_team','away_team']).\
    groupby(['pitch_type']).\
        agg(pd.Series.nunique)
```

```python
judge_df.filter(['pitch_type','home_team','away_team']).\
    groupby(['pitch_type']).\
        agg(pd.Series.nunique).\
            reset_index()
```
